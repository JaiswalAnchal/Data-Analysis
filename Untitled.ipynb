```python
import pandas as pd
import numpy as np
df=pd.read_excel('Book1.xlsx', index_col=0)
print(df.head(5))
```

```python
df.describe(include="all")
```

```python
df.dropna(subset=['Rainfall','Production'],axis=0,inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Predicted_Crop']=df.Crop.astype("category").cat.codes
print(df.head(10))
```

```python
freq= df['Predicted_Crop'].unique()
print(freq)
```

```python
Season = pd.get_dummies(df['Season'])
df = pd.concat([df,Season],axis = 1)
df.drop(["Season"],axis=1,inplace = True)
print(df.head(10))
```

```python

df=df.reset_index()
```

```python
print(df.head(10))
```

```python
X= df.drop(['Predicted_Crop','Crop'],axis=1) 
y= df['Predicted_Crop']
```

```python
print(X.head(10))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 8)
knn.fit(X_train, y_train)
```

```python
print("Train Score is {}".format(knn.score(X_train,y_train)))
print("Test Score is {}" .format(knn.score(X_test,y_test)))
```

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt

k_range = range(1,20)
scores = []

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(X_train, y_train)
    scores.append(knn.score(X_test, y_test))

plt.figure()
plt.xlabel('k')
plt.ylabel('accuracy')
plt.scatter(k_range, scores)
plt.xticks([0,5,10,15,20]);
```

```python
from sklearn.linear_model import LogisticRegression     
clf_log= LogisticRegression().fit(X_train,y_train)
print("Train Score is {}".format(clf_log.score(X_train,y_train)))
print("Test Score is {}".format(clf_log.score(X_test,y_test)))
```

```python
from sklearn.tree import DecisionTreeClassifier  
clf=DecisionTreeClassifier(max_depth=22).fit(X_train,y_train)
print("Train Score is {}".format(clf.score(X_train,y_train)))
print("Test Score is {}".format(clf.score(X_test,y_test)))
```

```python
from sklearn.ensemble import RandomForestClassifier

ranclf =RandomForestClassifier(n_estimators=25, max_depth=None,min_samples_split=2, random_state=0)  
ranclf.fit(X_train,y_train)
acc_ranclf_train = round(ranclf.score(X_train, y_train) * 100, 2)
acc_ranclf_test = round(ranclf.score(X_test, y_test) * 100, 2)
print("Train Score is {}".format(acc_ranclf_train))
print("Test Score is {}" .format(acc_ranclf_test))
```
